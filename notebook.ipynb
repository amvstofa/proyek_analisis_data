```python
""" Packages/Library yang Digunakan """
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Kategori produk apa yang paling sering muncul?
#Bagaimana distribusi panjang nama, deskripsi, jumlah foto, dan berat produk?
```

```python
#Data wragling | membuat kolom
products_dt = "./data/products_dataset.csv"
products_dt = pd.read_csv(products_dt)
products_dt.head()
```

```python
# melihat info
products_dt.info() 
```

```python
#Assesing data |  hasil pengecekan missing value
products_dt.isna().sum()
```

```python
# Mendeteksi Duplikasi
print("Jumlah duplikasi: ", products_dt.duplicated().sum())
```

```python
#Claning data | mengidentifikasi apakah terdapat duplikasi pada sebuah products_dt
products_dt.dropna(subset=['product_category_name'], inplace=True)
```

```python
"""
Insight 1: Kategori produk dengan jumlah terbanyak menunjukkan kategori yang paling sering tersedia di dataset.
Hal ini dapat membantu untuk menentukan kategori produk populer yang menjadi fokus pemasaran. ".

Insight 2: Korelasi tinggi antara berat dan salah satu dimensi produk dapat menunjukkan hubungan langsung (produk lebih berat memiliki dimensi lebih besar)
"""
```

```python
category_counts = products_dt['product_category_name'].value_counts()
```

```python
print("Jumlah Produk per Kategori:")
print(category_counts)
```

```python
# Pertanyaan 1: Kategori Produk yang Paling Banyak Tersedia
# Menghitung frekuensi kategori produk
plt.figure(figsize=(21,5))
category_counts.plot(kind='bar', color='#185336')
plt.title('Jumlah Produk per Kategori')
plt.xlabel('Kategori Produk')
plt.ylabel('Jumlah Produk')
plt.xticks(rotation=85)  
plt.show()
```

```python
# **Pertanyaan 2: Hubungan antara Berat Produk dan Dimensi Produk**
# Memeriksa korelasi antara berat produk dan dimensi produk
correlation_matrix = products_dt[['product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']].corr()
```

```python
# Menampilkan matriks korelasi
print("Matriks Korelasi:")
print(correlation_matrix)
```

```python
#Membuat scatter plot untuk berat vs panjang
plt.figure(figsize=(8,6))
plt.scatter(products_dt['product_weight_g'], products_dt['product_length_cm'], alpha=0.5, color='green')
plt.title('Berat Produk vs Panjang Produk')
plt.xlabel('Berat Produk (gram)')
plt.ylabel('Panjang Produk (cm)')
plt.show()
```

```python
# Membuat scatter plot untuk berat vs lebar
plt.figure(figsize=(8,6))
plt.scatter(products_dt['product_weight_g'], products_dt['product_width_cm'], alpha=0.5, color='orange')
plt.title('Berat Produk vs Lebar Produk')
plt.xlabel('Berat Produk (gram)')
plt.ylabel('Lebar Produk (cm)')
plt.show()
```

```python
# Membuat scatter plot untuk berat vs tinggi
plt.figure(figsize=(8,6))
plt.scatter(products_dt['product_weight_g'], products_dt['product_height_cm'], alpha=0.5, color='blue')
plt.title('Berat Produk vs Tinggi Produk')
plt.xlabel('Berat Produk (gram)')
plt.ylabel('Tinggi Produk (cm)')
plt.show()
```
